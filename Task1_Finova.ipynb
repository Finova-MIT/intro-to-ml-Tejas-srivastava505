```python
import pandas as pd  #did numpy, pandas and matplotlib together
import matplotlib.pyplot as plt
data=pd.read_csv("Movies.csv")   #already created this for an assignment
data
```

```python
data["Rating"].replace([0,np.nan],data["Rating"].mean(),inplace=True)
data
```

```python
plt.figure(figsize=(16,6))
plt.scatter(data["Movie Name"], data["Rating"],color='red')
plt.plot(data["Movie Name"],data["Rating"])
plt.xlabel("Movie Name")
plt.ylabel("Rating")
plt.show()
```

```python
high_rating = data["Rating"].idxmax()
movie_name = data.loc[high_rating]
print("The highest rating movie( is:\n", movie_name['Movie Name'])
```

```python
import seaborn as sns
plt.figure(figsize=(16,8))
seadata=data["Rating"].reset_index()  #messing with palette and printing the value above the graph.
sns.barplot(data=data, x=data["Movie Name"], y=seadata["Rating"], palette='coolwarm')
for index, value in enumerate(data["Rating"]):
    plt.text(index, value + 0.1, f"{value:.1f}", ha='center', fontsize=10, fontweight='bold', color='black')
plt.xlabel("Movie Name")
plt.ylabel("Rating")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv("sentiment.csv",index_col=0)
data["Text"]=data["Text"].astype(str).str.strip().str.lower()  #cleaning data here
data["Sentiment"]=data["Sentiment"].astype(str).str.strip().str.lower()
data
    
```

```python
df_a=data[(data["Sentiment"]=="positive") | (data["Sentiment"]=="negative") | (data["Sentiment"]=="neutral")]["Country"].str.strip()
tab1=pd.crosstab(df_a,data["Year"])
sns.heatmap(tab1,annot=True, fmt='d', cmap='YlGnBu')   #plotting heatmap for the tab1 data
plt.gca().collections[0].colorbar.set_label('Frequency')
```
